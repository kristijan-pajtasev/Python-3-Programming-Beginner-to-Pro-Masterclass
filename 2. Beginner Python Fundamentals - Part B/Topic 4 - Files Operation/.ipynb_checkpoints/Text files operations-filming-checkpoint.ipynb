# TEXT FILES
- fopen() is the key function to handle files 
- open() function takes two parameters; filename, and mode.
- Modes for files opening: 

    - "r" - Read - Default value. Opens a file for reading, error if the file does not exist
    - "a" - Append - Opens a file for appending, creates the file if it does not exist
    - "w" - Write - Opens a file for writing, creates the file if it does not exist
    - "x" - Create - Creates the specified file, returns an error if the file exists

- files can be used in a binary or text mode

    - "t" - Text - Default value. Text mode
    - "b" - Binary - Binary mode (e.g. images)

```python
f = open("my_file.txt", "r")
print(f.read())
```

```python
f = open("my_file.txt", "r")
print(f.read(15))
```

```python
f = open("my_file.txt", "r")
print(f.readline())
```

```python
f.close()
```

```python
f = open("my_file.txt", "a")
f.write("I am learning how to handle files in python!")
```

```python
f.close()
```

```python
# Create a new file
f = open("my_file_new_.txt", "x")
f.write("I am happy learning python now!")
f.close()
```

```python
f = open("my_file.txt", "r")
for line in f:
    print(line)
```

```python
with open("my_file.txt", "r") as f:
    text_data = f.readlines()
```

```python
for line in text_data:
    words = line.split()
    print(words)
```
